```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Skewness

### Zero Skewness in Data (Mean = Median = Mode)

```python
data1 = np.random.normal(0,1,40000)
plt.hist(data1,bins = 160)
plt.show()
```

### Positive Skewness in Data (Mean > Median > Mode)

```python
data2 = np.random.exponential(scale=1, size=30000)
plt.hist(data2,bins = 160)
plt.show()
```

### Negative Skewness in Data (Mean < Median < Mode)

```python
data3 = 10 - np.random.exponential(scale=1, size=30000)
plt.hist(data3,bins = 160)
plt.show()
```

# Kurtosis

```python
from scipy.stats import kurtosis
```

```python
data = np.arange(0,10)
plt.hist(data,bins = 10)
plt.title(f"Kurtosis value = {round(kurtosis(data),3)}")
plt.show()
data = np.random.standard_normal(10000)
plt.hist(data,bins = 100)
plt.title(f"Kurtosis value = {round(kurtosis(data),3)}")
plt.show()
```

# Data Visualization

### Types
1) Histogram
2) Boxplots
3) Scatter plots
4) Line plots
5) Bar charts
6) Density plots/KDE(Kernel Density Estimate) plot
7) Heatmaps
8) Pairplots
9) Violin plots
10) Pie charts

## Histogram

```python
data = np.random.standard_normal(1000)
plt.hist(data,bins = np.arange(min(data),max(data)+2) - 0.5,rwidth = .9,color="skyblue",edgecolor = 'black')
plt.savefig("Histogram")
plt.show()
```

## Boxplot

```python
data = np.random.standard_normal((100,2))
plt.boxplot(data,vert = True)
plt.xticks(ticks = [1,2],labels = ["D1","D2"])
plt.grid()
plt.savefig("Boxplot")
plt.show()
```

## Scatter Plot

```python
x = np.arange(0,11)
y = x**2
plt.scatter(x,y,c = 'orange')
plt.savefig("Scatter plot")
plt.grid()
plt.show()
```

## Line plot

```python
x = np.arange(0,11)
y = [1,2,3,2,3,4,3,4,5,6,7]
plt.plot(x,y)
plt.savefig("Line plot")
plt.show()
```

## Bar charts

```python
x = np.array(["Football","Cricket","Badminton"])
y = np.array([20,15,30])
plt.bar(x,y,width = 0.3,align='center',color = "lightgreen",edgecolor = "black")
plt.show()
```

## Density/KDE plot

```python
data = np.random.standard_normal(1000)
data = pd.DataFrame(data)
data.plot(kind = 'kde')
plt.savefig("Density plot")
plt.show()
```

## Heatmaps

```python
n = 6
y = np.array([])
for i in range(n):
    for j in range(n):
        y = np.append(y,i+j)
y = y.reshape(n,n)
plt.imshow(y,cmap = "hot")
plt.savefig("Heatmap")
plt.show()
```

## Pairplots

```python
x = np.random.randint(0,10,1000)
y = (x-5)**2
z = y + 5
var = np.array([x,y,z])
var_names = np.array(["X","Y","Z"])
fig, axes = plt.subplots(3,3,figsize = (8,8))
for i in range(3):
    for j in range(3):
        ax = axes[i,j]
        if i == j:
            ax.hist(var[i],bins = np.arange(min(var[i]),max(var[i]) + 2) - .5,color = "orange",rwidth = .9,edgecolor = "black")
        else:
            ax.scatter(var[i],var[j],c = "purple",s = 10)
        if j == 0:
            ax.set_ylabel(var_names[i])
        if i == 2:
            ax.set_xlabel(var_names[j])
        ax.set_xticks([])
        ax.set_yticks([])
plt.savefig("Pair plots")
plt.show()
```

## Violin plot

```python
data4 = np.random.standard_normal(10000)

vp = plt.violinplot(dataset = data4,showextrema=True,showmeans=True,showmedians=True)
for part in vp['bodies']:
    part.set_facecolor("blue")
    part.set_edgecolor("black")
    part.set_alpha(0.8)
vp['cmeans'].set_color("red")
vp['cmedians'].set_color("yellow")
vp['cmaxes'].set_color("lightblue")
vp['cmins'].set_color("lightblue")
vp['cbars'].set_color("lightblue")

plt.show()
```

## Pie chart

```python
labels = ['Football','Badminton','Cricket']
data = [20,30,15]
plt.pie(x = data,labels = labels,autopct='%1.2f%%')
plt.show()
```
