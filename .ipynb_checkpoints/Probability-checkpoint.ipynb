```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
avg = []
for i in range(30000):
    rolls = np.random.randint(0,11,100)
    avg.append(round(rolls.mean(),2))
plt.title("Normal distribution of tossing a coin to obtain Heads")
plt.hist(avg,bins=400)
plt.xlabel("Probability of Heads")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
